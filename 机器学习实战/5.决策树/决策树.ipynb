```python
#加载数据
import pandas as pd
import numpy as np
data = pd.read_csv('iris_dataset.csv')
data.head(5)
 
```

```python
#也可以使用sklearn自带的数据集
from sklearn.datasets import load_iris
iris = load_iris()
# print(iris)
```

```python
#定义数据X,y
X = data.drop(['target'],axis=1)
y = data.loc[:,'target']
X.head(5),y.head(5)
```

```python
#建立模型（决策树）
from sklearn import tree
#构造决策树实例 #criterion='entropy' 使用ID3（信息增益）作为分裂评价指标 min_samples_leaf=5 代表树的最深层为5
dc_tree = tree.DecisionTreeClassifier(criterion='entropy',min_samples_leaf=5)
#训练决策树
dc_tree.fit(X,y)
```

```python
#做出预测结果
y_predict= dc_tree.predict(X)
# y_predict
```

```python
#测试数据
```

```python
#对模型做出评估（准确率）
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y,y_predict)
accuracy
```

```python
#数据可视化
%matplotlib inline
from matplotlib import pyplot as plt

fig  = plt.figure()
tree.plot_tree(
    dc_tree,
    feature_names=['sepal_length','sepal_width','petal_length','petal_width'],#不需要与data中的数据标签对应
    class_names=['setoa','versicolor','virginica'],
    filled = 'True'
)
```
